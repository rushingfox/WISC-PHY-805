## Plot 3 decaying sine waves and save it as a PDF

Normal sine wave functions should be $sin(\omega t) = sin(2\pi f t)$
Decaying wave should be multiplied by a decaying factor of $e^{-\frac{t}{tau}}$

```python
def decaying_sine_wave(t, tau, f, phi, A):
    return A * np.exp(-t / tau) * np.sin(2 * np.pi * f * t + phi)
```

```python
t = np.arange(0, 5, 0.01)
y1 = decaying_sine_wave(t, tau=1, f=1, phi=0, A=1)
y2 = decaying_sine_wave(t, tau=0.5, f=2, phi=np.pi/4, A=0.5)
y3 = decaying_sine_wave(t, tau=2, f=0.5, phi=np.pi/2, A=2)

plt.figure(figsize=(10, 6))
plt.plot(t, y1, label='tau=1, f=1, phi=0, A=1')
plt.plot(t, y2, label='tau=0.5, f=2, phi=π/4, A=0.5')
plt.plot(t, y3, label='tau=2, f=0.5, phi=π/2, A=2')
plt.title('Decaying Sine Waves')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.savefig('decaying_sine_waves.pdf', bbox_inches='tight', dpi=600, format='pdf')
plt.show()
```
